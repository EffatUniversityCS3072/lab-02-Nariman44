#### Practice Exercise 2
Read the coordinates of the capital cities of the world from the country-capital-lat-long-population.csv file. Use NumPy to print the name and coordinates of the capital city closest to the Saudi Arabia capital - Riyadh.

Note that:

1. The Country Name for Saudi Arabia is given as Saudi Arabia in the data.
2. The ‘closeness’ of capital cities from the Saudi Arabia capital is based on the Euclidean distance of their coordinates to those of the Saudi Arabia capital.


Hints:

1. Use read_html() from the Pandas library to read the table.
2. Use the to_numpy() function of the Pandas DataFrame class to convert a DataFrame to a Numpy array
3. Use broadcasting to compute the euclidean distance of capital cities from Riyadh.

```python
import pandas as pd
import numpy as np

# Assuming data has been provided as a dictionary
data = {
    'Country': ['Afghanistan', 'Albania', 'Saudi Arabia', 'Algeria', 'Argentina', 'Australia', 'Austria', 'Bahrain', 'Bangladesh'],
    'Capital City': ['Kabul', 'Tirana', 'Riyadh', 'Algiers', 'Buenos Aires', 'Canberra', 'Vienna', 'Manama', 'Dhaka'],
    'Latitude': [34.5289, 41.3275, 24.6905, 36.7525, -34.6051, -35.2835, 48.2064, 26.2154, 23.7104],
    'Longitude': [69.1725, 19.8189, 46.7096, 3.042, -58.4004, 149.1281, 16.3707, 50.5832, 90.4074]
}

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Get coordinates of Riyadh
riyadh_lat, riyadh_lon = df[df['Capital City'] == 'Riyadh'][['Latitude', 'Longitude']].values[0]

# Convert latitude and longitude columns to NumPy arrays
latitudes = df['Latitude'].to_numpy()
longitudes = df['Longitude'].to_numpy()

# Compute the Euclidean distance from Riyadh to all other capitals
distances = np.sqrt((latitudes - riyadh_lat)**2 + (longitudes - riyadh_lon)**2)

# Set distance of Riyadh to itself as infinity to exclude it
distances[df['Capital City'] == 'Riyadh'] = np.inf

# Find the index of the minimum distance
min_distance_idx = np.argmin(distances)

# Get the closest capital city information
closest_city = df.iloc[min_distance_idx]

# Output the result
print("Closest capital city to Riyadh:")
print(f"City: {closest_city['Capital City']}")
print(f"Latitude: {closest_city['Latitude']}")
print(f"Longitude: {closest_city['Longitude']}")





```
